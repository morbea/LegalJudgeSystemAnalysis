```python
def mark_vice_ministers(df):
    # Initialize the 'had_vice' column to 0
    df['had_vice'] = 0
    vice_ministers_index = df.index[df["משרד"] == "סגני השרים"].tolist()
    if vice_ministers_index:
        index1 = vice_ministers_index[0]
        vice_ministries = df["משרד"].iloc[index1 + 1:].unique()
        df.loc[df["משרד"].isin(vice_ministries), "had_vice"] = 1
        df = df.iloc[:index1]

    return df

# df = mark_vice_ministers(df)
# df
```

```python
def shift_row_right(row):
    """
    Shift all elements of the row one position to the right.
    The first element of the row will become NaN.
    """
    return [None]  + row[:-1].tolist()

def merge_offices_with_shared_ministers(df):
    """
    Merge offices that shared ministers by shifting rows to the right
    when the 'משרד' column is NaN.
    """
    for i in range(len(df)):
        if pd.isna(df.loc[i, 'משרד']):
            df.iloc[i] = shift_row_right(df.iloc[i])
    return df

# df1= merge_offices_with_shared_ministers(df)
# df1
```

```python
def rename_columns(df: pd.DataFrame):
    old_columns = ['מפלגה', 'שרי הממשלה', 'סגני השרים', 'dates']
    new_columns = ['שם', 'מפלגה', 'תאריך עברי', 'תאריך לועזי']
    return df.rename(columns = dict(zip(old_columns, new_columns)))
```

```python
def fill_office(df: pd.DataFrame):
    """
    If office is empty, take the office of the previous row 
    (wiki merge cells are making problems so if there is no office, we assume it is the office above)
    """
    # Find a start and an end index to start "rolling" the rows values (only if minister office has an empty value)
    start_idx = df.columns.get_loc('תאריך לועזי')
    end_idx = df.columns.get_loc('שר')
    for i, row in df.iterrows():
        if i >= 2:
            if not row['משרד'] or pd.isna(row['משרד']):
                    for j in range(start_idx, end_idx, -1):
                        col_name = df.columns[j]
                        prev_col_name = df.columns[j-1]
                        df.at[i, col_name] = df.at[i, prev_col_name]
                    df.at[i, 'משרד'] =  df.at[i-1, 'משרד']
    return df
```

```python
def fill_na_in_df(df: pd.DataFrame):
    df.fillna(method='ffill', inplace=True)
    df.replace('', np.nan, inplace=True)  # Replace empty strings with NaN
    df.fillna(method='ffill', inplace=True)
```

```python
def detect_invalid_rows(df:pd.DataFrame) -> pd.DataFrame:
    """
    Detectes invalid rows in a data frame - rows where name contains digits
    """

    return df[~df['שם'].str.contains(r'\d')]
```

```python
def process_df(df: pd.DataFrame):
    df = mark_vice_ministers(df)
    df = merge_offices_with_shared_ministers(df)
    df.drop(0, inplace=True)  # First row is always invalid
    df = rename_columns(df)
    df = fill_office(df)
    # Drop unnecessary columns
    df.drop(columns=['שר','כהונה'], inplace=True)
    fill_na_in_df(df)
    df = detect_invalid_rows(df)
    return df
```
